### DAY 6: PROMPT ENGINEERING

```python
from dotenv import load_dotenv
from langchain_mistralai import ChatMistralAI
import pandas as pd

load_dotenv()
```

```python
model = ChatMistralAI(
    model="mistral-small-2506",
    temperature=0.3
)

def run_prompt(prompt):
    response = model.invoke(prompt)
    return response.content


test_email = """
Hello Support Team,

My account has been hacked, and I can no longer log in.

Someone has changed my email address and made unauthorized purchases using my saved payment method.

Please help me recover my account immediately.

Thank you.
"""
```

```python
# PROMPT STRATEGIES

strategies = {

    "Zero Shot":
    """
Classify the following customer support email.

Return only:

Category:
Reason:

Email:

{input}
""",

    "One Shot":
    """
Example:

Email:
My payment failed but money was deducted.

Category:
URGENT

Reason:
Financial issue requiring immediate attention.

--------------------------------

Now classify the following.

Email:

{input}

Category:
""",

    "Few Shot":
    """
Example 1

Email:
My account has been hacked.

Category:
URGENT

Reason:
Security issue.

----------------------------

Example 2

Email:
Can you send your product brochure?

Category:
NOT_URGENT

Reason:
General inquiry.

----------------------------

Example 3

Email:
How do I update my profile picture?

Category:
NOT_URGENT

Reason:
Routine support request.

----------------------------

Now classify.

Email:

{input}

Category:
""",

    "Role Prompt":
    """
You are a Senior Customer Support Manager.

Your responsibility is to prioritize support tickets.

Analyze the email and provide:

Category:
Reason:

Email:

{input}
""",

    "Chain of Thought":
    """
Think step by step.

1. Identify the customer's issue.

2. Determine its impact.

3. Determine urgency.

Finally provide:

Category:
Reason:

Email:

{input}
""",

    "ReAct":
    """
You are an expert customer support analyst.

Observe:
Identify important facts.

Reason:
Explain the urgency.

Act:
Provide the final classification.

Email:

{input}

Observe:
"""

}
```

```python
import os
from dotenv import load_dotenv

load_dotenv()

MISTRAL_API_KEY = os.getenv("MISTRAL_API_KEY")
```

```python
model = ChatMistralAI(
    model="mistral-small-2506",
    api_key=MISTRAL_API_KEY,
    temperature=0.3
)
```

```python
results = []

for strategy_name, prompt_template in strategies.items():

    print(f"Executing: {strategy_name}")

    prompt = prompt_template.replace("{input}", test_email)

    output = run_prompt(prompt)

    results.append({
        "Prompt Version": strategy_name,
        "Input": test_email.strip(),
        "Output": output,
        "Notes": ""
    })

    print("Completed ✔")


# CREATE DATAFRAME

df = pd.DataFrame(results)

# DISPLAY RESULTS

print("\n")
print("=" * 80)
print("PROMPT OUTPUTS")
print("=" * 80)

for index, row in df.iterrows():

    print("\n")
    print("-" * 80)

    print("Strategy :", row["Prompt Version"])

    print("\nOutput:\n")

    print(row["Output"])

# SAVE CSV

df.to_csv("prompt_testing_results.csv", index=False)

print("\nResults saved as prompt_testing_results.csv")

# SIDE-BY-SIDE COMPARISON

print("\n")
print("=" * 100)
print("SIDE-BY-SIDE COMPARISON")
print("=" * 100)

comparison = df[["Prompt Version", "Output"]]

print(comparison.to_string(index=False))



print("\n")

print("Day 6 complete!")

print("Key insight: test prompts systematically, one change at a time.")

print("Consistency across runs matters more than a single impressive output.")

print("\nDeliverables Completed:")
```
