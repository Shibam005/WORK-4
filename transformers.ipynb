### TRANSFORMERS

```python
from transformers import AutoTokenizer, pipeline    # Libraries imported 
```

### TASK 1: TOKENIZATION OF A PARAGRAPH

```python
print("\n")
print("=" * 70)
print("TASK 1 : TOKENIZATION")
print("=" * 70)

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

paragraph = """
Hello I am Shibam Pal , a student of Ramco Institute of technology . 
I am currently pursuing BTech degree in Artificial Intelligence and data science.
I have a great passion for evolving modern technology mainly in field of AI."""

encoding = tokenizer.encode(paragraph, add_special_tokens=True)
tokens = tokenizer.convert_ids_to_tokens(encoding)

print("\nOriginal Paragraph:\n")
print(paragraph)

print("\nToken IDs:\n")
print(encoding)

print("\nTokens:\n")
print(tokens)

print("\nTotal Tokens:", len(tokens))

```

### TASK 2: SUMMARIZATION OF NEWS

```python
print("\n")
print("=" * 70)
print("TASK 2 : NEWS SUMMARIZATION")
print("=" * 70)

news_article = """
The transformer architecture was introduced in 2017 and rapidly
changed the field of Natural Language Processing. Unlike recurrent
neural networks, transformers process all words simultaneously using
a mechanism called self-attention. This enables them to capture
relationships between distant words more effectively while training
faster on modern hardware. Today, transformer-based models power
chatbots, translation systems, summarizers, and many state-of-the-art
AI applications.
"""
# Model 1 : T5
summarizer_t5 = pipeline(
    "summarization",
    model="t5-small"
)

summary1 = summarizer_t5(
    news_article,
    max_length=60,
    min_length=20,
    do_sample=False
)[0]["summary_text"]

print("\nSummary using T5-small:\n")
print(summary1)


# Model 2 : BART

summarizer_bart = pipeline(
    "summarization",
    model="facebook/bart-large-cnn"
)

summary2 = summarizer_bart(
    news_article,
    max_length=60,
    min_length=20,
    do_sample=False
)[0]["summary_text"]

print("\nSummary using BART-large-CNN:\n")
print(summary2)

```

### TASK 3: REVIEW TEXT CLASSIFICATION

```python
# TASK 3 : TEXT CLASSIFICATION
print("\n")
print("=" * 70)
print("TASK 3: TEXT CLASSIFICATION - REVIEW SENTIMENT ANALYSIS")
print("=" * 70)


classifier = pipeline(
    "sentiment-analysis",
    model="distilbert-base-uncased-finetuned-sst-2-english"
)

reviews = [

    "This HP Victus laptop I am using  is really good and amazing",

    "Worst purchase ever. Completely disappointed.",

    "The movie was average and not very exciting.",

    "Excellent course with detailed explanations.",

    "The service was terrible and slow."
]

for review in reviews:

    result = classifier(review)[0]

    print("\nReview:")
    print(review)

    print("Prediction:", result["label"])

    print("Confidence:", round(result["score"], 4))

```

### TASK 4: TRANSLATION OF TEXT

```python

# TASK 4 : TRANSLATION

print("\n")
print("=" * 70)
print("TASK 4 : ENGLISH TO FRENCH TRANSLATION")
print("=" * 70)

english_sentence = (
    "Transformer models have revolutionized natural language processing."
)

# Translation Model 1

translator1 = pipeline(
    "translation_en_to_fr",
    model="t5-small"
)

translation1 = translator1(english_sentence)[0]["translation_text"]

print("\nTranslation using T5-small:\n")
print(translation1)

# Translation Model 2

translator2 = pipeline(
    "translation_en_to_fr",
    model="Helsinki-NLP/opus-mt-en-fr"
)

translation2 = translator2(english_sentence)[0]["translation_text"]

print("\nTranslation using Helsinki-NLP:\n")
print(translation2)

print("Comparison:")
print("- T5 performs multiple NLP tasks using one model.")
print("- Helsinki-NLP is specialized for translation and usually produces")
print("  more natural and accurate French sentences.")

```

###  Encoder vs Decoder: My Comparison

**Encoder models** are mainly designed to **understand** the input text. They read the entire sentence at once and learn the relationships between all the words. Because of this, encoder models such as **BERT** are very effective for tasks like sentiment analysis, text classification, and question answering.

**Decoder models**, on the other hand, are designed to **generate** text. They produce one token at a time based on the previous tokens. Models such as **GPT** are well suited for tasks like text generation, story writing, and dialogue generation.

From my observations, the **encoder "feels" like a reader that analyzes and understands the meaning of the text**, while the **decoder "feels" like a writer that generates new text based on what it has already seen**.

In summary:

* **Encoder → Understands text**
* **Decoder → Generates text**
* **Encoder-Decoder → First understands the input and then generates an output, making it suitable for tasks like summarization and translation.**
